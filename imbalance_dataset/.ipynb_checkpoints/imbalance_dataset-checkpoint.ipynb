```python
in classification, imbalance dataset is a problem, imbalance dataset yani output m koi catagory
bhhtt ziada a jati h , is s y problem ati h k hmara dataset bias ho jata h yani vo just usi catagory pr
train ho jata h jo catagory ziada hoti h output m:
    e.g: agr hm n model ko train krvaya 500 data points pr jis m 400 ki output cat thi and 100 ki output dog thi to model
bias ho jay ga or cat valy ko mutabik train ho jay ga phir testing k time jab hm us ko dog ki bhi picture dikhay gy
vo phir bhi output m cat hi de ga
```

```python
there are 2 techiques for handling imbalance data:
                     1- sample under sampling
                     2-sample over sampling
```

<h1>random under sampling

```python
is m hm vo catagory jo majority m h us k samples minority k samples k equal krty hn
----> yani agr cat 400 h and dog is 100 (as of above example) to hm cat k 100 samples uthay gy randomly
and new dataset vo dataset ho ga jis m 100 cat and 100 dog ho ga
```

<h1>sample over sampling

```python
is s dog valy 100 ko 400 k equal kry gy q k cat 400 h, yani dog valy 100 ko repeat krty rhy gy bar bar jab tak
400 na ho jay 
```

<h1> practice question with biased data

```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
dataset=dataset.drop(columns=['User ID','Gender'], axis=1)
dataset.head()
```

```python
dataset['Purchased'].value_counts()  #to yani data imbalance h yani baised h, phly is baised pr hi 
# model bna kr dekhty hn k kia accuracy ati h
```

```python
x=dataset.drop(columns='Purchased')
y=dataset['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
# 50	20000	1
d=np.array([50,	20000])
lr.predict([d])     #dekho ab us n zero dia q k model biased tha 0 pr due to its high number
```

<h1> practice question with biased data convert to unbiased

```python
import pandas as pd
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df=df.drop(columns=['User ID','Gender'], axis=1)
df.head()
```

```python
x=df.drop(columns='Purchased')
y=df['Purchased']
```

```python
random over samplin ya random under sampling dono k liye hmy imblearn(I am balanced learn) library ko
install krna h via pip
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus=RandomUnderSampler()
x_new, y_new=rus.fit_resample(x,y)    #similary you can do for RamdomOverSampling too
```

```python
y_new.value_counts()   
```

```python
y_new
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_new, y_new, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
# 50	20000	1
d=np.array([50,	20000])
lr.predict([d])    # to dekho ab us n correct 1 dia, yani ab model unbaised h
```
