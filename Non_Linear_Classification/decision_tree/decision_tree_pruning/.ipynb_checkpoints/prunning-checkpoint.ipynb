```python
pruning means cutting branches from tree, it is done to 
1- avoid over fitting
2- to increase accuracy 
```

```python
how to check overfitting?
    agar hmary model ki training ki accuracy and testing ki accuracy m bhhtt difference h to yani k hmara 
model over fitted h, training ki accuracy ziada h and testing ki kam h
        ----> prunning hmesha nhi krni, jab bhi model over fitted ho tab hi prunning krni h
```

```python
There are 2 ways of prunning:
    1- pre prunning
    2- post prunning
Pre Prunning: Train krny s phly hi pruning krni h
Post Prunning: Train krny k ander different levels ko dety rehty hn
```

# pre prunning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
dataset=dataset.drop(columns=['User ID','Gender'], axis=1)
dataset.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='Age', y='EstimatedSalary', data=dataset, hue='Purchased')
plt.show()
```

```python
x=dataset.drop(columns='Purchased', axis=1)
y=dataset['Purchased']
```

# post prunning

```python
from sklearn.tree import DecisionTreeClassifier
i=1
while(i<10):
    dtc_post_prun=DecisionTreeClassifier(max_depth=i)
    dtc_post_prun.fit(x_train, y_train)
    print(f'max_depth is: {i}')
    print('training score is',dtc_post_prun.score(x_train, y_train)*100)
    print('testing score is',dtc_post_prun.score(x_test, y_test)*100)
    print("\n")
    print("\n")
    i=i+1


# ab sab s avhi accuracy or over fitting s avoidance max_depth 4 pr mil rhi , is liye hm 4 max_Depth hi final kryn gy, or
# is pr model ko train kr k decision regions bna len gy
```

```python
dtc_post_prun=DecisionTreeClassifier(max_depth=4)
dtc_post_prun.fit(x_train, y_train)
```
