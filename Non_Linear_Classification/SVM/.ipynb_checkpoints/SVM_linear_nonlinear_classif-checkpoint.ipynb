```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('placement-dataset.csv')
dataset=dataset.drop(columns=['Unnamed: 0'], axis=1)
dataset.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='cgpa', y='iq', data=dataset, hue='placement')
plt.show()
```

```python
x=dataset.drop(columns='placement', axis=1)
y=dataset['placement']
```

```python
from sklearn.preprocessing import kernals
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=30)
```

```python
from sklearn.svm import SVC
# 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'
svm=SVC(kernel='linear')
svm.fit(x_train, y_train)
# linear is good
```

```python
svm.score(x_train, y_train)*100, svm.score(x_test, y_test)*100
# score is very low without kernals, we will use kernals to increase score
# random_state = 3 is providing best!
```

```python
from mlxtend.plotting import plot_decision_regions
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=svm)
plt.show()
```
