```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset = dataset.drop(columns=['User ID','Gender','Age'], axis=1)
dataset.head()
```

```python
plt.figure(figsize=(3,2))
sns.scatterplot(x='EstimatedSalary', y='Purchased', data=dataset)
plt.show()
```

```python
x=dataset.drop(columns=['Purchased'], axis=1)
y=dataset['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train), lr.score(x_test, y_test)
```

```python
plt.figure(figsize=(3,2))
sns.scatterplot(x='EstimatedSalary', y='Purchased', data=dataset)
# plt.plot(x, lr.predict(x), color='red')
sns.lineplot(x='EstimatedSalary', y=lr.predict(x), data=dataset, color='red')
plt.show()
```
