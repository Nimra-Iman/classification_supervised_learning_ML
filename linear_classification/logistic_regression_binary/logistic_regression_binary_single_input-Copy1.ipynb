```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset=dataset.drop(columns=["User ID","Gender","EstimatedSalary"], axis=1)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# first check, whether linear relationship or not:
plt.figure(figsize=(5,3))
sns.scatterplot(x="Age", y="Purchased", data=dataset)
plt.show()
# yani logistic regression can be applied
```

```python
x=dataset[["Age"]]
y=dataset[["Purchased"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train, y_train) 
```

```python
lr.score(x_test, y_test)
```

```python
# 26	0
predicted = lr.predict(x_test)
predicted
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Age", y="Purchased", data=dataset)
sns.lineplot(x="Age",y=lr.predict(x), data=dataset, color="red")
plt.show()
```
