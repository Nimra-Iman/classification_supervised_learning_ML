```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset=pd.read_csv('mushroom_cleaned_polynomial (1).csv')
dataset = dataset.drop(columns = ["cap-shape", "gill-attachment", "gill-color", "season", "stem-width", "stem-height"])
dataset.head()
```

```python
x=dataset.drop(columns=['class'], axis=1)
y=dataset['class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.3, random_state=42)
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)     #data is not linearly separable , if we use use logostic regression directly, without adding
# polynomial features, look, the score is less,,,,,,,  we must have to add polynomial features to it 
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='cap-diameter', y='stem-color', data=dataset, hue='class')
plt.show()
# as you can see data is not linearly separable, so will add polynomial features to dataset
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=2)
pf.fit(x)
x=pd.DataFrame(pf.transform(x))
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```
