<h2> types of Naive Baise (NB) algo

```python
1- Gaussian NB : suitable for data havinf columns have normal/gaussain distribution
2- Bernouli NB : suitable for data having columns have binary form, either 0 and 1
3- Multinomial NB: suitable for data having columns have discrete data such as text data
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv('placement-dataset.csv')
dataset=dataset.drop(columns=['Unnamed: 0'], axis=1)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
x=dataset.drop(columns=['placement'])
y=dataset['placement']
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='cgpa', y='iq', data=dataset, hue='placement')
plt.show()   #yes, linearly separable
```

```python
# whether there is normal distribution or not
plt.figure(figsize=(4,3))
sns.histplot(dataset['cgpa'], kde=True)
plt.show()   #almost normal
```

```python
plt.figure(figsize=(4,3))
sns.histplot(dataset['iq'], kde=True)
plt.show()   #almost normal
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
# data is normally distributed, so we will use gaussian NB
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
# we will use GaussianNB here, BernoulliNB and MultinomialNB are imported just to show hpw to use it
```

```python
gnb=GaussianNB()
gnb.fit(x_train, y_train)
```

```python
gnb.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(4,3))
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb)
plt.show() 
```

```python
bnb=BernoulliNB()
bnb.fit(x_train, y_train)
bnb.score(x_test, y_test)*100 #of course score is less, because this dataset is not suitable for BernoulliNB
```

```python
mnb=MultinomialNB()
mnb.fit(x_train, y_train)
mnb.score(x_test, y_test)*100 #of course score is less, because this dataset is not suitable for MultinomialNB
```
